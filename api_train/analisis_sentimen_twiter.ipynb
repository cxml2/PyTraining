```python
import tweepy
import re
from textblob import TextBlob 
```

```python
API_KEY="your api key"
API_SECRET_KEY="your secret"
B_TOKEN="not use this for now"
ACCESS_TOKEN="your token"
ACCESS_SECRET_TOKEN="your secret"
```

```python
auth = tweepy.OAuthHandler(API_KEY,API_SECRET_KEY)
auth.set_access_token(ACCESS_TOKEN, ACCESS_SECRET_TOKEN)
api = tweepy.API(auth)
```

```python
# cari=api.search(q="gila",lang="en", count=10)

# # cari
```

```python
cari2=api.search(q="wendy red velvet",lang="en", count=10)
for tweet in cari2:
  tp = {}
  tp['tanggal_tweet']=tweet.created_at
  tp['pengguna']=tweet.user.screen_name
  tp['Isi tweet']=tweet.text
  #tclean untuk bersihkan datanya
  t_clean =' '.join(re.sub("(@[A-Za-z0-9]+)|([^0-9A-Za-z \t])|(\w+:\/\/\S+)"," ",tweet.text).split())
  analisis=TextBlob(t_clean)
  # print(analisis.sentiment.polarity)
  if analisis.sentiment.polarity > 0.0:
    tp['sentimen'] = "positif"
  elif analisis.sentiment.polarity == 0.0:
    tp['sentimen'] = "netral"
  else:
    tp['sentimen'] = "negatif"
  print(tp)

  # last video 15:30
```
