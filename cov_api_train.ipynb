```python
import json
import requests 

#UJI KONEKSI SEBELUM PEMANGGILAN API
url_1="https://data.covid19.go.id/public/api/update.json"
url_2="https://coronavirus-19-api.herokuapp.com/all"
url_3="https://covid19.mathdro.id/api"
url_4="https://dekontaminasi.com/api/id/covid19/stats"
# url = url_2
def con(url):
  timeout = 1
  try:
	  request = requests.get(url, timeout=timeout)
	  print("Connected to the Internet")
  except (requests.ConnectionError, requests.Timeout) as exception:
	  print("No internet connection.")


print("API Kemkes 1 : ")
con(url_1)
print("API Inter : ")
con(url_2)
print("API MathDroid : ")
con(url_3)
print("API Dekontaminasi")
con(url_4)


```

```python
import json
import requests as r
from datetime import datetime

#API Dekontaminasi
web=r.get("https://dekontaminasi.com/api/id/covid19/stats")
dat=web.json()
print(dat)
print("")
print("Negara :",dat['name'])
print("- Kasus Terkonfirmasi :",dat['numbers']['infected'])
print("- Sembuh :",dat['numbers']['recovered'])
print("- Meninggal :",dat['numbers']['fatal'])
print("")
x=dat['timestamp']
print("- Updated on :",x,"\n")
# print(dat['regions'][1])

x=len(dat['regions'])
print("data provinsi")
for i in range(x):
  print(i+1,"-\nProvinsi :",dat['regions'][i]['name'])
  print("Total Kasus : ",dat['regions'][i]['numbers']['infected'])
  print("Total Sembuh : ",dat['regions'][i]['numbers']['recovered'])
  print("Total Meninggal : ",dat['regions'][i]['numbers']['fatal'],"\n")
```
